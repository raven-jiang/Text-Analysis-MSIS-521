```python
import nltk
import pandas as pd
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
review = pd.read_csv('IA2.csv', header =None)
review.head()
```

```python
r = review.iloc[:,1]
r.head()
```

```python
#Step 1 Tokenizing
tokens = []
for text in r:
    tokenize = nltk.word_tokenize(text.lower())
    tokens.append(tokenize)
print(tokens)
```

```python
#Remove Whitespace

t_white =[]
for text in r:
    tokenize_w = nltk.tokenize.WhitespaceTokenizer()
    tw = tokenize_w.tokenize(text.lower())
    t_white.append(tw)
print(t_white)
```

```python
#Step 2 Lemmatize

lemmatize =[]
lemmatizer = nltk.stem.WordNetLemmatizer()
for text in t_white:
    lemmatized_token = [lemmatizer.lemmatize(token) for token in text 
                        if token.isalpha()]
    lemmatize.append(lemmatized_token)
print(lemmatize)
```

```python
# Step 3 remove stop words
from nltk.corpus import stopwords

stop_words_removed = []
for text in lemmatize:
    line = [token for token in text if not token in stopwords.words('english') 
            if token.isalpha()]
    stop_words_removed.append(line)
print(stop_words_removed)
```

```python
# low frequency words
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# change into strings for Vectorizer
s=[]
for line in stop_words_removed:
    strings = ' '.join(line)
    s.append(strings)
```

```python
#bag of 3 grams
vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=3)
vectorizer.fit(s)
v = vectorizer.transform(s)
```

```python
print(v.toarray())
```

```python
print(vectorizer.vocabulary_)
```

```python
#Step 5 Pos Tag
POS_TAG = []
for text in tokens:
    POS = nltk.pos_tag(text)
    
    POS_token_temp = []
    for i in POS:
        POS_token_temp.append(i[0] + i[1])
    POS_TAG.append(" ".join(POS_token_temp))
    
    print(POS_TAG)

#pos tag vectorize
v_postag = TfidfVectorizer (min_df=4)
v_postag.fit(POS_TAG)
v2 = v_postag.transform(POS_TAG)

print(v2.toarray())

print(v_postag.vocabulary_)
```

```python
df2 = pd.DataFrame(v2.A)
df2.head()
```
